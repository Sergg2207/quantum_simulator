# Simulación: Primer sistema (Sección 4.1)

Sistema: partícula confinada a N posiciones discretas en una línea. Definimos kets por amplitudes en cada posición.

```python
import numpy as np
from quantum_utils import normalize, prob_position
```

## Definición del sistema

N = 7

ket inicial: amplitudes donde la partícula está principalmente en la posición 3

```python
N = 7
psi = np.zeros(N, dtype=complex)
psi[3] = 1.0
psi = normalize(psi)
print('Ket inicial (normalizado):', psi)
for i in range(N):
    print(f'P(pos={i}) = {prob_position(psi, i):.4f}')
```

## Ejemplo: cambiar número de posiciones y estado inicial

```python
N = 5
psi2 = normalize([1, 1, 0, 0, 0])
print('psi2:', psi2)
print('P(pos=0)=', prob_position(psi2, 0))
```
